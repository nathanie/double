```python
from __future__ import print_function
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
card_labels = pd.read_clipboard()
```

```python
card_labels = card_labels.T.fillna(-1)
```

```python
card_labels.columns = card_labels.iloc[0,:]
card_labels.drop('car_file_number',axis=0,inplace=True)
```

```python
card_labels[card_labels>-1] = 1
```

```python
card_labels[card_labels==-1] = 0
```

```python
card_labels.sort_index().to_csv('card_files_labels.csv')
```
